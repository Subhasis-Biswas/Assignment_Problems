#### $f(x)=x^3-0.165x^2+3.993\times 10^{-4}$

```python
#importing the necessary libraries
import matplotlib.pyplot as plt
import math

#Defining the function
def f(x):
    return (x**3-0.165*x**2+3.993*10**(-4)) #function



a,b=0,0.11  #initializing the interval, [a,b]
c=(a+b)/2  #initializing the midpoint, c
root_seq=list() #we're storing the approximation to the root at each stage in a list. "root_seq" stands for "root sequence".
root_seq+=[0.1,c] #the first root_seq is 0.1, and the second root_seq is the midpoint, c. It has been done to avoid the division by zero error, 
               #or the "list index out of range" error in the while loop

tolerable_error=10**(-4)
iteration_count=1        #The first midpoint is already calculated, so we start the iteration count from 1.

err_seq=list()
err_seq+=[abs(root_seq[-1]-root_seq[-2])/abs(root_seq[-1])]   #Storing the relative error at each stage in a list. "err_seq" stands for "error sequence".
                                                              #err_seq[-1] is the last element of the list, and err_seq[-2] is the second last element of the list.

while(err_seq[-1]>tolerable_error):  #The iteration scheme for the bisection method.
    if f(a)*f(c)<0:
        b=c
    else:
        if(f(c)==0):                 #We break the loop if the midpoint is a root.
            break
        else:
            a=c
    c=(a+b)/2
    root_seq+=[c]
    err_seq+=[abs(root_seq[-1]-root_seq[-2])/abs(root_seq[-1])]    #add the new relative error to the list of errors
    iteration_count+=1

print(f'The root is {c} and the relative error is {err_seq[-1]}, with {iteration_count} iterations')
print(f'The root rounded off to 4 significant figures is {round(c,5)}')   #The decimal place is upto 5, but the significant figures are 4, due to the presence of a leading zero.

fig,axs=plt.subplots(1,2,figsize=(20,5))
axs[0].plot(root_seq,'o-')
axs[0].set_title('Root Approximation Sequence')
axs[0].set_xlabel('Iteration Number')
axs[1].plot(err_seq,'o-',color='red')
axs[1].set_title('Relative Error Sequence')
axs[1].set_xlabel('Iteration Number')
plt.show()
```
