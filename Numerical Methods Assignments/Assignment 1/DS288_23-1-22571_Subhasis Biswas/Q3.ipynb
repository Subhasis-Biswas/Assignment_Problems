## Problem 1

$f(x)=(x+e^{-x^2}\cos{x})^2$\
$f'(x)=2e^{-2x^2}(e^{x^2}x+\cos{x})(e^{x^2}-\sin{x}-2x\cos{x})$\
$f''(x)=2(-e^{-x^2}\sin{x}-2e^{-x^2}x\cos{x}+ 1)^2 + 2(x + e^{-x^2}\cos{x})(4e^{-x^2}x\sin{x}+4e^{-x^2}x^2\cos{x}-3e^{-x^2}\cos{x})$


### Performing Newton-Raphson with $x_0=0$ and with relative error tolerance $10^{-6}$:

```python
#Importing the necessary libraries

import math
import matplotlib.pyplot as plt


#Defining the functions through return

def f_1(x):    #The function itself
    return (x+math.e**(-x**2)*math.cos(x))**2

def Df_1(x):   #The first derivative of the function
    return 2*(x+(math.exp(-1*x**2))*math.cos(x))*(1-math.sin(x)*math.exp(-1*x**2)-2*x*math.cos(x)*math.exp(-1*x**2))



n=0   #Initialising the number of iterations
x=[0]  #The initial guess
x+=[x[n]-f_1(x[n])/Df_1(x[n])]  #The first iteration of Newton's method

n+=1 #The number of iterations

error=list() #The list of relative errors
error+=[abs(x[n]-x[n-1])/abs(x[n])] #The first relative error

while(error[n-1]>10**(-6)):           #The Iteration Scheme for Newton's method
    if(f_1(x[n])==0):                 #If the function is zero, the iteration stops
        break   
    x+=[x[n]-f_1(x[n])/Df_1(x[n])]
    n+=1
    error+=[abs(x[n]-x[n-1])/abs(x[n])]

print(f'The root of the function is {x[n]} with relative error {error[n-1]} and number of iterations {n}')
print(f'The root rounded to 6 decimal places is {round(x[n],6)}')

error_1=error #Saving the relative errors for plotting later
m=n #Saving the number of iterations for plotting later

#Plotting the relative error against the number of iterations
plt.plot(range(1,n+1),error,'-o',color='blue')
plt.xlabel('Number of iterations')
plt.ylabel('Relative error')
plt.xticks(list(range(1,n+1)))
plt.title('Relative error against number of iterations in Newton\'s method')
plt.show()
```

### Performing Modified Newton's Method with $x_0=0$ and relative error tolerance $10^{-6}$:

## Problem 2:

$f(x)=x-2\log(1+e^{-x})$\
$f'(x)=\displaystyle\frac{1+3e^{-x}}{1+e^{-x}}$\
$f''(x)=\displaystyle\frac{-2e^x}{(1+e^x)^2}$

### Performing Newton's Method with $x_0=0$ and relative error tolerance $10^{-6}$:


```python
#Importing the necessary libraries

import math
import matplotlib.pyplot as plt


#Defining the functions through return
def f_2(x):   #The function itself
    return (x-2*math.log(1+math.e**(-x)))    
def Df_2(x):  #The first derivative
    return (1+3*math.e**(-x))/(1+math.e**(-x))


n=0    #initialising the number of iterations
x=[0]  #The initial guess
x+=[x[n]-f_2(x[n])/Df_2(x[n])]
n+=1   #Incrementing the number of iterations
error=[abs(x[n]-x[n-1])/abs(x[n])]

while(error[n-1]>10**(-6)):                #The Iteration Scheme for Newton's method
    if(f_2(x[n])==0):                      #If the function is zero, the iteration stops
        break
    x+=[x[n]-f_2(x[n])/Df_2(x[n])]
    n+=1
    error+=[abs(x[n]-x[n-1])/abs(x[n])]

print(f'The root of the function is {x[n]} with relative error {error[n-1]} and number of iterations {n}')
print(f'The root rounded to 6 decimal places is {round(x[n],6)}')


error_1=error #Saving the relative errors for plotting later
m=n #Saving the number of iterations for plotting later

#Plotting the relative error against the number of iterations
plt.plot(range(1,n+1),error,'-o',color='blue')
plt.xlabel('Number of iterations')
plt.ylabel('Relative error')
plt.xticks(list(range(1,n+1)))
plt.title('Relative error against number of iterations in Newton\'s method')
plt.show()
```

### Performing Modified Newton's Method with $x_0=0$ and relative error tolerance $10^{-6}$:


```python
#Importing the necessary libraries

import math
import matplotlib.pyplot as plt


#Defining the functions through return
def f_2(x):      #The function
    return (x-2*math.log(1+math.e**(-x)))
def Df_2(x):     #First derivative of the function
    return (1+3*math.e**(-x))/(1+math.e**(-x))
def Gf_2(x):     #The Second derivative of the function
    return -(2*math.e**x)/(math.e**x + 1)**2


n=0 #Initialising the number of iterations
x=[0]  #The initial guess
x+=[x[n]-f_2(x[n])*Df_2(x[n])/(Df_2(x[n])**2-f_2(x[n])*Gf_2(x[n]))] #The first iteration of Newton's method

n+=1 #Increase the number of iterations done

error=list() #The list of relative errors
error+=[abs(x[n]-x[n-1])/abs(x[n])] #The first relative error

while(error[n-1]>10**(-6)):                  #The Iteration Scheme for Modified Newton's method
    if(f_2(x[n])==0):                        #If the function is zero, the iteration stops
        break
    x+=[x[n]-f_2(x[n])*Df_2(x[n])/(Df_2(x[n])**2-f_2(x[n])*Gf_2(x[n]))]
    n+=1
    error+=[abs(x[n]-x[n-1])/abs(x[n])]

print(f'The root of the function is {x[n]} with relative error {error[n-1]} and number of iterations {n}')
print(f'The root rounded to 6 decimal places is {round(x[n],6)}')
error_2=error
m=n

#Plotting the relative error against the number of iterations
plt.plot(range(1,n+1),error,marker='o',color='red',linestyle='dashed')
plt.xlabel('Number of iterations')
plt.ylabel('Relative error')
plt.xticks(list(range(1,n+1)))
plt.title('Relative error against number of iterations in Modified Newton\'s method')
plt.show()
```
